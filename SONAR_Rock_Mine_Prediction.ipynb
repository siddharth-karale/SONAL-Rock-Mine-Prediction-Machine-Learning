```python
import pandas as pd #importing pandas
import numpy as np #importing numpy
import seaborn as sns #importing seaborn
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
sonar = pd.read_csv("K:\DATA SCIENCE\DataSets\Sonar data.csv",header= None)
```

```python
X = sonar.drop(60, axis=1)
Y = sonar[60]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
pred_model = model.predict(X_test)
```

```python
confusion_matrix(Y_test, pred_model)
```

```python
cm = accuracy_score(Y_test, pred_model)
```

```python
print('Accuracy of the model on training data:', cm*100,'%')
```
